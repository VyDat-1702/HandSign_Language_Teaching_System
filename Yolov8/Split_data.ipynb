```python
import os
import shutil
import random

# Đường dẫn gốc
ROOT = 'D:/install/AI/Code/Project/yolo/V8/Yolov8/data_collect'
IMAGE_DIR = os.path.join(ROOT, 'images')
LABEL_DIR = os.path.join(ROOT, 'labels')
OUTPUT_DIR = os.path.join(ROOT, 'split_data')  # thư mục kết quả

# Danh sách các class
classes = [
    "A", "B", "C", "D", "DD", "E", "G", "H", "I", "K",
    "L", "M", "N", "O", "P", "Q", "R", "S", "T", "U",
    "V", "X", "Y"
]

# Tỷ lệ chia
ratios = {'train': 0.7, 'val': 0.2, 'test': 0.1}

# Tạo thư mục đầu ra (train/val/test/images + labels)
for split in ratios:
    os.makedirs(os.path.join(OUTPUT_DIR, split, 'images'), exist_ok=True)
    os.makedirs(os.path.join(OUTPUT_DIR, split, 'labels'), exist_ok=True)

# Duyệt qua từng class
for cls in classes:
    # Lấy danh sách ảnh theo tiền tố class (ví dụ: A1.jpg, A2.jpg,...)
    images = [f for f in os.listdir(IMAGE_DIR) if f.startswith(cls) and f.endswith(('.jpg', '.png'))]
    random.shuffle(images)

    total = len(images)
    train_end = int(ratios['train'] * total)
    val_end = train_end + int(ratios['val'] * total)

    split_data = {
        'train': images[:train_end],
        'val': images[train_end:val_end],
        'test': images[val_end:]
    }

    for split, files in split_data.items():
        for img_name in files:
            label_name = img_name.rsplit('.', 1)[0] + '.txt'

            # Copy image
            shutil.copy(
                os.path.join(IMAGE_DIR, img_name),
                os.path.join(OUTPUT_DIR, split, 'images', img_name)
            )

            # Copy label
            label_path = os.path.join(LABEL_DIR, label_name)
            if os.path.exists(label_path):
                shutil.copy(
                    label_path,
                    os.path.join(OUTPUT_DIR, split, 'labels', label_name)
                )
            else:
                print(f"⚠️ Label file not found for: {img_name}")

    print(f"✅ Đã chia xong class {cls}")

print("🎉 Tất cả class đã được chia hoàn tất!")
```
