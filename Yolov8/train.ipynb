```python
import cv2
from ultralytics import YOLO

# Load model YOLOv8
model = YOLO("D:/install/AI/Code/Project/yolo/V8/Yolov8/runs/detect/train/weights/best.pt")

# Má»Ÿ camera (0 lÃ  camera máº·c Ä‘á»‹nh)
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("KhÃ´ng thá»ƒ má»Ÿ camera!")
    exit()

while True:
    # Äá»c tá»«ng khung hÃ¬nh tá»« camera
    ret, frame = cap.read()
    if not ret:
        print("KhÃ´ng thá»ƒ Ä‘á»c khung hÃ¬nh!")
        break
    img = cv2.flip(frame, 1)
    # Dá»± Ä‘oÃ¡n vá»›i YOLO
    results = model(frame)  # Truyá»n khung hÃ¬nh vÃ o mÃ´ hÃ¬nh

    # Hiá»ƒn thá»‹ hÃ¬nh áº£nh vá»›i cÃ¡c há»™p dá»± Ä‘oÃ¡n
    # Äáº£m báº£o báº¡n truy cáº­p Ä‘á»‘i tÆ°á»£ng hÃ¬nh áº£nh chÃ­nh xÃ¡c trong káº¿t quáº£
    frame_with_boxes = results[0].plot()  # Láº¥y Ä‘á»‘i tÆ°á»£ng Ä‘áº§u tiÃªn trong káº¿t quáº£ vÃ  váº½ cÃ¡c há»™p

    # Hiá»ƒn thá»‹ hÃ¬nh áº£nh
    cv2.imshow("YOLO Detection", frame_with_boxes)  # Hiá»ƒn thá»‹ hÃ¬nh áº£nh vá»›i cÃ¡c há»™p dá»± Ä‘oÃ¡n

    # Náº¿u nháº¥n 'q', thoÃ¡t khá»i vÃ²ng láº·p
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Giáº£i phÃ³ng camera vÃ  Ä‘Ã³ng cá»­a sá»•
cap.release()
cv2.destroyAllWindows()
```
