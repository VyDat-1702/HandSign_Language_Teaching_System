```python
import cv2
from ultralytics import YOLO

# Load model YOLOv8
model = YOLO("D:/install/AI/Code/Project/yolo/V8/Yolov8/runs/detect/train/weights/best.pt")

# Mở camera (0 là camera mặc định)
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Không thể mở camera!")
    exit()

while True:
    # Đọc từng khung hình từ camera
    ret, frame = cap.read()
    if not ret:
        print("Không thể đọc khung hình!")
        break
    img = cv2.flip(frame, 1)
    # Dự đoán với YOLO
    results = model(frame)  # Truyền khung hình vào mô hình

    # Hiển thị hình ảnh với các hộp dự đoán
    # Đảm bảo bạn truy cập đối tượng hình ảnh chính xác trong kết quả
    frame_with_boxes = results[0].plot()  # Lấy đối tượng đầu tiên trong kết quả và vẽ các hộp

    # Hiển thị hình ảnh
    cv2.imshow("YOLO Detection", frame_with_boxes)  # Hiển thị hình ảnh với các hộp dự đoán

    # Nếu nhấn 'q', thoát khỏi vòng lặp
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Giải phóng camera và đóng cửa sổ
cap.release()
cv2.destroyAllWindows()
```
